### Check model assumptions

To justify using simple linear regression, check that the four linear regression assumptions are not violated. These assumptions are:

- **Linearity:** There is a linear relationship between the independent variable(s) and the dependent variable. This means that the relationship between the variables can be represented by a straight line. You can check this assumption by creating a scatter plot of the data and visually inspecting if the points roughly form a linear pattern.

- **Independence:** The observations in the dataset are independent of each other. This assumption is crucial because violating it can lead to biased and unreliable estimates. In simple terms, the value of one data point should not be influenced by the value of another data point.

- **Normality of Residuals:** The residuals (the differences between the observed and predicted values) are normally distributed. Checking the normality assumption typically involves creating a quantile-quantile (QQ) plot of the residuals and assessing whether the points roughly follow a straight line.

- **Homoscedasticity:** The residuals have constant variance across the range of predicted values. In other words, the spread of the residuals should be roughly the same for all predicted values. You can examine this assumption by creating a scatter plot of the residuals against the predicted values.

### Model assumption 3: Normality of Residuals

The normality assumption states that the errors are normally distributed.

Create two plots to check this assumption:

* **Plot 1**: Histogram of the residuals
* **Plot 2**: Q-Q plot of the residuals

```python
# Calculate the residuals.
residuals = model.resid

# Create a 1x2 plot figures.
fig, axes = plt.subplots(1, 2, figsize = (10,4))

# Create a histogram with the residuals. 
sns.histplot(residuals, ax=axes[0])

# Set the x label and title of the residual plot.
axes[0].set_xlabel("Residual Value")

# Set the title of the residual plot.
axes[0].set_title("Histogram of Residuals")

# Create a Q-Q plot of the residuals.
sm.qqplot(residuals, line='s', ax=axes[1])

# Set the title of the Q-Q plot.
axes[1].set_title("Q-Q Plot of Residuals")

# Use matplotlib's tight_layout() function to add space between plots for a cleaner appearance.
plt.tight_layout()
```

### Model assumption 4: Homoscedasticity

The **homoscedasticity (constant variance) assumption** is that the residuals have a constant variance for all values of `X`.

Check that this assumption is not violated by creating a scatterplot with the fitted values and residuals. Add a line at $y = 0$ to visualize the variance of residuals above and below $y = 0$.

```python
# Create a scatterplot with the fitted values from the model and the residuals.

sns.scatterplot(x=model.fittedvalues, 
                y=model.resid)

# Set the x-axis label.
plt.xlabel("Fitted Values")

# Set the y-axis label.
plt.ylabel("Residuals")

# Set the title.
plt.title("Fitted Values versus Residuals")

# Add a line at y = 0 to visualize the variance of residuals above and below 0.

plt.axhline(0, color='r') 

plt.tight_layout()
```

**Note in `sns.scatterplot(x=model.fittedvalues, y=model.resid)`:**
- `x`: This refers to the predicted (fitted) values of the dependent variable (Y) based on the linear regression model. The `model.fittedvalues` contains the predicted values of the dependent variable for each data point in the dataset. These predicted values are calculated by the linear regression model based on the independent variable(s) (X) and the estimated coefficients.

- `y`: This represents the residuals of the linear regression model. The residuals are the differences between the observed values of the dependent variable and the corresponding predicted (fitted) values. Mathematically, the residuals (e) are calculated as:
        
        e = observed value of Y in ['Sales'] - predicted value of Y in ['model.fittedvalues']

### Measure the uncertainty of the coefficient estimates

Model coefficients are estimated. This means there is an amount of uncertainty in the estimate. A p-value and $95\%$ confidence interval are provided with each coefficient to quantify the uncertainty for that coefficient estimate.

Display the model results again.

```python
# Display the model_results defined previously.

model_results
```

**Question:** Based on this model, what is your interpretation of the p-value and confidence interval for the coefficient estimate of X?

**Answer:**

The p-value of the model is 0.000, which is very small. This means that the probability of obtaining the observed results by chance is very low. This suggests that the independent variable, `TV`, is likely to have a significant effect on the dependent variable, `Sales`.

 
The confidence interval with 95% confidence level is [3.558, 3.565]. This means that we are 95% confident that the true value of the beta coefficient lies within this interval. In other words, we can be 95% confident that for every one million increase in `TV` variable, the dependent variable `Sales` is expected to increase by between 3.558 and 3.565 million, holding all other independent variables constant (if any -- in this case there is no other independent variables).
